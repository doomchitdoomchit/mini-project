### 3.Network Core

- Packet Switching(패킷 교환)
  - Entire data broken into small packets
  - Each packet has its destination address
  - Each packet handled independently
  - Packet transmitted at full link capacity
  - Takes L/R seconds to transmit L-bit packet into at R bps
  - Store-and-forward : entire packet must arrive at router before forwarded to the next
  - End-end delay := 2L/R

### 5.Performance Metrice
성능지표
 - Delay(지연시간)
  - Packet delivering time from source to destination
  - $d_{nodal} = d_{proc} + d_{queue} + d_{trans} + d_{prop}$
    - proc(processing delay) : 패킷이 도달 했을 때, '라우터에서 패킷을 처리하기 위한 주소가 무성이며 이주소 따라서 어디로 보내야 하는가' 라는 것을 평가
    - queue(queuing delay)* : 어떤 패킷이 도달했을 때, 이전에 미리 도착한 패킷이 있어다면 큐에 들어가서 대기를 해야함.
    - trans(transmission delay)* : 링크의 전송률에 따라서 패킷 사이즈가 굉장히 크다면 걸리는 시간
    - prop(propagation delay) : 전파시간, 전파 딜레이
  - *표시는 중요하다는 뜻
    - queueing delay는 버퍼에서 내가 얼마나 기다리는가, 중간 라우터에서 도착해서 보내지기 까지 거기서부터 다음 hop으로 전달 되기까지 버퍼에서 얼마나 기다리는가 하는 것
    - 라우터가 격는,라우터에 얼마나 많은 데이터가 도달해 있는가 라는 것에 의존적
    - 사용자의 행태하고도 관련이 있기 때문에 정확하게 얼마 만큼이 걸릴 것이다하는 보장이 없음.
    - transmission delay는 패킷의 length와 링크의 밴드 width, 링크의 capacity에 따라서 계산
    - $d_{trans} = L/R$ (L : packet length(bits), R : link bandwidth(bps)
  - 나머지
    - processing delay는 캐쉬에 도착 했을 때 비트에러가 포함이 되어 있느냐 라는 것을 결정해야 하고 목적자 주소를 보고 아웃풋 링크를 결정해야 하는 일들이 일어나지만 요즘 굉장히 하드웨어 속도가 빨라 젺기 때문에 이런 작업은 빠른 속도로 이루어질 수 있고 하드웨어 적으로 처리하게 되면 굉장히 빠른 속도로, 하나하나의 패킷을 처리하는 속도는 무시할 수 있을 정도의 값이기 때문에 그렇게 중요하지 않다.
    - propagation delay는 비트단위로 빛의 속도의 2/3정도로 보낸다. 그러나 물리적인 거리에 의해 보내는 시간이 100ms정도가 되면 데이터 상으로는 상당히 오랜 시간이 걸린다고 볼 수 있다. 그러나 목적지가 정해지고 나서부터는 꼬리를 물고 지연 없이 보내지기 때문에 propagation delay는 제일 초기의 지연시간이고 그 뒤로는 큰 영향이 없어서 중요하게 보지 않는다.
 - Packet Loss(손실률)
  - Ratio of lost packets to total sent packets
  - PDR(Packet Delivery Ratio) : Ratio of packets successfully deliverd to the total number of packets
  - 네트워크에서 패킷로스는 중간의 라우터에서 보통 일어남. 라우터에서 갑자기 버퍼가 저장할 수 있는 공간, 양이 있는데 이것을 넘어서는 데이터가 한번에 도착하게 되면 용량을 넘어서는 패킷들은 잃어버림
 - Throughput(전송률)
  - The amount of traffic delivered per time
  - instantaneous(순간) : 가장 피크일 때, 어느정도의 스루풋이냐 하는 것
  - average(평균) : 어떤 전체시간을 보고 평균을 가지고 계산
  - bottleneck link : 두 소스와 목적지 사이에 링크 중에 전송용량이 낮은 링크가 전체 패스의 용량을 결정

###2.Web and HTTP

HTTP Overveiw
  - HTTP (HyperText Transfer Protocol)
    - Web`s application payer protocol
    - Hyperlink : a reference to data the reader can directly follow by clicking
  - client/server model
    - client : browser that requests, recives, and "displays" Web objects
    - server : web server sends objects in respense to requests
  - Based on TCP
    1. Client initiates TCP connection (creates socket) to server
    2. Server accepts TCP connection from client
    3. HTTP messages(application-layers protocol messages) exchanged between browser (HTTP client) and Web server(HTTP server)
    4. TCP connection closed

  - RTT : Round-Trip-Time
  - HTTP response time
    - one RTT to initate TCP connection
    - one RTT for HTTP request and first few bytes of HTTP response to return
    - file transmission time
    - Total response time = 2RTT + file transmission time

HTTP Version
  - HTTP/1(1996)
    - Non-persistent HTTP: one object per TCP connection
    - long latency
  - HTTP/1.1(1999)
    - Presistent HTTP: multiple objects over one TCP connection
    - Lagency decrease
    - Synchronous order of response/requesr pairs over one TCV connection
  - HTTP/2(2015)
    - Persistent HTTP
    - Asynchronous(parallel) multiple request/response pairs over one connection
    - Compatible with HTTP/1.1

REST
  - REpresentational State Transfer
    - HTTP should be 'stateless', proposed by PhD. Roy Fielding in 2000
    - Used in designing the HTTP/1.1 and Uniform Resource Indentified(URI) standard
    - RESTful service : service that conforms to the REST architectural style

  - Architectural Constraints
    - client-server architecture : seperation of the user interface concerns form the data storage concerns
    - statelessness : no client context being stored on the server between requests
    - cacheability : server responses cacheable on client and intermediaries
    - layered system : unable to tell whether a client is directly connected to the end server or to an intrtmediary along the way
    - code on demand (optional) : able to transfer executable code such as Java applets and Java script
    - uniform interface : simplification and decoupling of the architecture

###7.Peer-to-Peer Aplication

P2P Architecture
- No always-on server
- Arbitrary end systems directly communicate
- Peers are intermittently connected and change IP addresses
- example:
  - file distribution (BitTorrnet)
  - streaming (KanKan)
  - Voip (Skype)

Case Study : BitTorrnet
- File divide into 256 kb chunks
- Peers in torrent send/receive file chunks
- Chunk receiving
  - at any given time, different peers have different subsets of file chunks
  - periodically, Alice asks each peer for list of chunks that they have
  - Alice requests missing chunks from peers, rarest first

- Free-rider
  - peer usually wants to receive file without sending file to others

- Chunk sending : tit-for-tat
  - Alice sends chunks to those four peers currently sending her chunks at hightest rate
    - other peers are choked by Alice (do not receive chunks from her)
    - re-evaluate top 4 every 10 secs
  - every 30 secs : randomly select another peer, starts sending chunks
    - 'optimistically unchoke' this peer
    - newly chosen peer may join top 4

BitTorrent : tit-for-tat
  1. Alice 'optimistically unchokes' Bob
  2. Alice becomes one of Bob's top-four providers; Bob resiprocates
  3. Bob becomes one of Alice's top-four providers

###3.User Data Protocol

User Datagram Protocol [RFC 768]
- 'No frills', 'bare bones' Internet transport protocol
- Connectionless service:
  - each UDP segment handled independently of others
  - Unreliable : UDP segments may be lost or delivered out-of-order to app

- UDP use:
  - streaming multimedia apps (loss tolerant, rate sensitive)
  - DNS
  - SNMP
- Reliavle transfer over UDP:
  - add reliability at application layer
  - application-specific error recovery!

UDP Segment Header
- Advantages of UDP
  - no connection establishment (which can add delay)
  - simple : no connection state at sender, receiver
  - small header size
  - no congestion control : UDP can blast away as fast as desired

UDP Checksum
- Sender
  - make a 16-bit integer checksim code of segment contents including header
  - put the checksum into the checksum field and transmit the segment

- Receiver
  - compute checksum of received segment
  - check if the computed checksum equals all 1's or not

Stop-and-Wait
- stop-and-wait
  - sender sends one packet, the waits for receiver respone
  - after recieving ACK, sender resumes transmission
  - if timer expires without receiving ACK, sender retransmits the previous packet

Pipelined Protocols
- pipelining : sender allows multiple, 'in-flight', yet-to-be-acknowledged packets

Pipelined Protocols : Overview
- Go-back-N

- Selective repeat
  - Sender can have up to N unpacked packets in pipline
  - Reciever sends individual ack for each packet
  - Sender has timer for each unacked packet

  


##WEEK5 Transmission Control Protocol

###5.Transmission Control Protocol

TCP Overview
- point-to-point : one sender, one reciever
- connection-oriented service
  - reliable transfer, in-order delivery
  - handshaking initializes sender and receiver state before data exchange
- Pipelined transmission
  - window size is set by TCP congestion and flow control
  - congestion control : transmission rate controlled not to make congestion in network
  - flow control : sender will not overwhelm receiver
  - MSS : maximum segment size
  - Full-deplex connection
    - bi-directional data flow in same connection

Sequence and Acknowledgment Number
- Sequence number
  - byte stream 'number' of first byte in segment`s data
- Acknowledgment number
  - sequence number of the next segment expected by receiver
  - cumulative ACK

TCP Reliable Data Transfer
- TCP create reliable data transfer service on top of IP's unreliable service
  - pipelined segments
  - cumulative acks
  - single retransmission timer

- TCP sender 
  - when receiving data from app
    - create segment with seq. number
    - send it and start timer if not already running
  - when timer expires
    - retransmit the seqment that caused timeout and restart timer
  - when receiving ack
    - update ACKed packet list
    - start timer if there are still unacked segments

- TCP receiver
  - when receiving data from sender
    - if discover bit error through checksum, then drop the packet
    - checking seq. # and there being no gap, send cumulative ACK
    - if packet is duplicated, drop it
- How receiver handle out-of-order segments?
  - TCP specification does not say explicitly-up to implementer

Fast Retransmit
- Timeout period often relatively long : 
  - long delay before resending lost packet
- Another way to detect segment lost
  - whenever a segment is received. receiver sends ACK to request the segment that has the next sequence number of the last in-order segment
  - sender often sends many segments back-to-back
  - if a segment is lost, there will likely be many duplicate ACKs
- TCP fast retransmit
  - if sender receivers 3 duplicate ACKs(expect the first normal ACK) for same data, resend unpacked segmaent with smallest seq. number

- As a result, TCP retransmissions triggersd by:
  1. timeout events
  2. duplicate acks

TCP Flow Control   
Receiver controls sender, so sender won't overflow receiver's buffer by transmittion too much, too fast
- Reciver advertise free buffer space by including rwnd value in TCP header of receiver-to-sender segment
  - RcvBuffer size set via socket options(typical default is 4096 bytes)
  - many OSs autoadjust RcvBuffer
- Sender limits amount of unpacked ('inflight') data to receiver's rwnd value
- Guarantees receive buffer will not overflow

###6.Congestion Control

Network Congestion   
Congestion
- Too many source sending too much data too fast for network to handle
- Different from flow control
  - congestion control is global issue
  - flow control is point-to-point issue
- Manifestations:
  - lost packets (buffer overflow at routers)
  - long delays (queueing in router buffers)

Congestion Control Approaches
- Two approaches towards congestion control
- end-to-end
  - No explicit feedback from network
  - Congestion inferred from end-system observed loss, delay
  - Approach taken by TCP
- network-assisted
  - Routers provide feedback to end systems
    - single bit indicating congestion (SNA, DECbit, TCP/IP ECN, ATM)
    - explicit rate for sender to send at

TCP Congestion Control Overview
- AIMD approach : sender increase transmission rate( window size), probing for usable bandwidth, until loss occurs
  - additive increase : increase cwnd by 1 MSS every RTT until loss detected
  - multiplicative decrease : cut cwnd in hall after loss
- Congestion window(cwnd)
  - the rate a TCP sender can send traffic (he amount of packets in transit)
  - LastBySent - LastByteAcked <= min(cwnd, rwnd)
- cwnd is dynamic:
fuction of perceived network congestion

Change of cwnd : Slow Start
- When connection begins, increase rate exponentially until first loss event:
  - initially cwnd = 1 MSS
  - double cwnd every RTT
  - done by incrementing cwnd for every ACK received
- Summary : initial rate is slow but ramps up exponentially fast

Change of cwnd : Congestion Avoidance
- Q : When should the exponrntial growth in the slow-start state end?
- A : When cwnd gets to 1/2 of its value before congestion was detected

- Implementation
  - variable ssthresh
  - on loss event, ssthresh is set to 1/2 of cwnd just before loss event
  - surpassing ssthresh, cwnd increases by just a single MSS every RTT


###7.TCP Congestion Control Algorithm

TCP Tahoe(1988)
- Features
  - slow start
  - congestion avoidance
  - fast retransmit
    - retransmission upon 3 duplicate ACKs
  - The slow start begins ( i.e. cwnd = 1 MSS) on
    - timeout
    - fast retransmit

TCP Reno(1990)
- Limitation of Tahoe
  - no differentiates losses indicated by timout and duplicate ACKs
  - always set cwnd to 1 MSS
- Two response to packet losses by Reno
  - loss indicated by timeout
    - cwnd set to 1 MSS
    - slow start to ssthresh, then congestion avoidance begins
    - loss indicated by fast retransmit
      - cwmd cut in half
      - fast recovery, then congestion avoidance

Fast Recovery
- Half cwnd maintained until a non-duplicate ACK is received
  - amount of packets in transit kept from decreaseing more than expected
- Operation
  - Upon the event of 3 duplicate ACKs
  - after reducing cwnd to 1/2
  - the algorithm not only retransmits the oldest unacknowledged packet,
  - but also inflates the congestion window by the number of duplicate packets
  - A non-duplicate ACK make the congestion avoidance start.

TCP NewReno(1999)
- Vulnerability of Reno's fast recovery
  - in case of multiple packet losses, cwnd may br reduce too much(exponential way)

- Refinement of Reno's fast recovery
  - resistant to multiple losses
  - restricts the exit from the recovery phase until all data packets from the initial congestion window are acknowledged
  - exit from the NewReno's fast recovery only when a new data ACK is received
  - Keeps the sequence number of the last data packet sent before entireing the fast recovery state

###8.TCP vs. UDP

TCP vs. UDP
- TCP fairness through AIMD (additive increase and multiplicative decrease)
  - two competing sessions
- Multimedia apps often do not use TCP
